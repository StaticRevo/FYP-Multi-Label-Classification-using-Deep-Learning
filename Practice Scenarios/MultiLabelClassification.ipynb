### Customised Pytorch Dataset Class

```python
class Dataset(object):
    def __init__(self, index):
        raise NotImplementedError
    def __len__(self):
        raise NotImplementedError
    def __getitem__(self, index):
        raise NotImplementedError
    def __add__(self, other):
        return ConcatDataset([self, other])
```

```python
class MultiLabelDataset(Dataset):
    def __init__(self, image_size):
        self.x_train, self.x_val, self.x_test, self.y_train, self.y_val, self.y_test = None, None, None, None, None, None
        self.mode = 'train'

        self.labels, self.images = [], []

        labels_path = r'C:\Users\isaac\Downloads\multi-label dataset\multilabel_modified\labels.csv'
        images_path = r'C:\Users\isaac\Downloads\multi-label dataset\multilabel_modified\images'

        df = pd.read_csv(labels_path)
        column_names = df.columns.tolist()
        df.rename(columns={column_names[1]: 'Classes'}, inplace=True)

        for index, row in df.iterrows():
            l = list(row.iloc[-10:])
            self.labels.append(l)
        self.labels = np.array(self.labels)

        image_names = df['Image_Name'].tolist()
        for img_name in tqdm(image_names):
            image_path = os.path.join(images_path, img_name)
            image = Image.open(image_path)
            image = np.array(image)
            if len(image.shape) == 2:
                # Convert grayscale image to 3-channel image
                image = np.stack([image] * 3, axis=-1)
            image = cv2.resize(image, (64, 64))
            image = image.reshape((3, 64, 64)) # Channels, Height, Width
            self.images.append(image)

        self.images = np.array(self.images)
        self.normalize()

    def normalize(self):
        self.images = self.images / 255.0

    def train_val_test_split(self):
        self.x_train, self.x_test, self.y_train, self.y_test = train_test_split(self.images, self.labels, test_size=0.33, random_state=42)
        self.x_train, self.x_val, self.y_train, self.y_val = train_test_split(self.x_train, self.y_train, test_size=0.33, random_state=42)

    def __len__(self):
        if self.mode == 'train':
            return self.x_train.shape[0]
        elif self.mode == 'val':
            return self.x_val.shape[0]
        elif self.mode == 'test':
            return self.x_test.shape[0]

    def __getitem__(self, index):
        if self.mode == 'train':
            sample = {'images': self.x_train[index], 'labels': self.y_train[index]}
        elif self.mode == 'val':
            sample = {'images': self.x_val[index], 'labels': self.y_val[index]}
        elif self.mode == 'test':
            sample = {'images': self.x_test[index], 'labels': self.y_test[index]}

        return sample
```

### Model Building

```python
class MLC(nn.Module):
    def __init__(self, num_classes):
        super(MLC, self).__init__()
        self.resnet = models.resnet18(weights=ResNet18_Weights.DEFAULT)
        self.in_features = self.resnet.fc.in_features
        self.resnet.fc = nn.Linear(self.in_features, num_classes)

    def forward(self, x):
        x = self.resnet(x)
        return x
```

### Training Loop

```python
image_size = (128, 128)
mlc_dataset = MultiLabelDataset(image_size=image_size)
mlc_dataset.train_val_test_split()
```

```python
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
dataloader = DataLoader(mlc_dataset, batch_size=32, shuffle=True)
num_classes = 10
model = MLC(num_classes)
model = model.to(device)

eta = 0.00005
num_epochs = 10
optimizer = torch.optim.Adam(model.parameters(), lr=eta)

epoch_train_loss = []
mlc_dataset.mode = 'train'
```

```python
# Assuming you have already defined your model, dataloader, optimizer, and device
num_epochs = 10
epoch_train_loss = []
epoch_train_accuracy = []

for epoch in range(num_epochs):
    train_losses = []
    correct_predictions = 0
    total_predictions = 0
    model.train()
    
    for D in dataloader:
        optimizer.zero_grad()
        data = D['images'].to(device, dtype=torch.float)
        labels = D['labels'].to(device, dtype=torch.float)
        y_hat = model(data)
        
        # Calculate loss
        error = nn.BCEWithLogitsLoss()
        loss = torch.sum(error(y_hat, labels))
        loss.backward()
        optimizer.step()
        train_losses.append(loss.item())
        
        # Calculate accuracy
        predictions = torch.sigmoid(y_hat) > 0.5
        correct_predictions += (predictions == labels).sum().item()
        total_predictions += labels.numel()
    
    epoch_train_loss.append(np.mean(train_losses))
    epoch_train_accuracy.append(correct_predictions / total_predictions)
    
    print(f'Epoch {epoch + 1}/{num_epochs}, Loss: {epoch_train_loss[-1]}, Accuracy: {epoch_train_accuracy[-1]}')
```

```python
# Plotting the metrics
plt.figure(figsize=(10, 5))

# Plot loss
plt.plot(epoch_train_loss, label='Train Loss', color='red')
plt.xlabel('Epoch')
plt.ylabel('Loss', color='red')
plt.tick_params(axis='y', labelcolor='red')

# Create a second y-axis to plot accuracy
ax2 = plt.gca().twinx()
ax2.plot(epoch_train_accuracy, label='Train Accuracy', color='blue')
ax2.set_ylabel('Accuracy', color='blue')
ax2.tick_params(axis='y', labelcolor='blue')

# Add legends
plt.legend(loc='upper left')
ax2.legend(loc='upper right')

plt.title('Training Loss and Accuracy')
plt.grid()
plt.show()
```
